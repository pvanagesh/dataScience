```python
import pandas as pd
```

```python
dfMovies=pd.read_csv(r'D:\Nagesh\DataScience\ClassNotes\6th\movie.csv')
dfMovies.head()
```

```python
dfMovies['year'] = dfMovies['title'].str.extract('.*\((.*)\).*', expand=True)
dfMovies
```

```python
dfMovies['Name']=dfMovies['title'].str[:-6]
dfMovies
```

```python
movie_genres = dfMovies['genres'].str.split('|', expand=True)
movie_genres[:10]
```

```python
dfMovies['isAdventure']=dfMovies['genres'].str.contains('Adventure')
dfMovies
```

```python
dfMovies['isAnimation']=dfMovies['genres'].str.contains('Animation')
dfMovies['isChildren']=dfMovies['genres'].str.contains('Children')
dfMovies['isComedy']=dfMovies['genres'].str.contains('Comedy')
dfMovies['isFantasy']=dfMovies['genres'].str.contains('Fantasy')
dfMovies['isHorror']=dfMovies['genres'].str.contains('Horror')
dfMovies['isThriller']=dfMovies['genres'].str.contains('Thriller')
dfMovies['isRomance']=dfMovies['genres'].str.contains('Romance')
dfMovies['isDrama']=dfMovies['genres'].str.contains('Drama')
dfMovies['isCrime']=dfMovies['genres'].str.contains('Crime')
dfMovies
```
