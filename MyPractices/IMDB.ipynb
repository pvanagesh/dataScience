```python
dfMovies['year'] = dfMovies['title'].str.extract('.*\((.*)\).*', expand=True)
dfMovies['Name']=dfMovies['title'].str[:-6]
dfMovies['isAdventure']=dfMovies['genres'].str.contains('Adventure')
dfMovies['isAnimation']=dfMovies['genres'].str.contains('Animation')
dfMovies['isChildren']=dfMovies['genres'].str.contains('Children')
dfMovies['isComedy']=dfMovies['genres'].str.contains('Comedy')
dfMovies['isFantasy']=dfMovies['genres'].str.contains('Fantasy')
dfMovies['isHorror']=dfMovies['genres'].str.contains('Horror')
dfMovies['isThriller']=dfMovies['genres'].str.contains('Thriller')
dfMovies['isRomance']=dfMovies['genres'].str.contains('Romance')
dfMovies['isDrama']=dfMovies['genres'].str.contains('Drama')
dfMovies['isCrime']=dfMovies['genres'].str.contains('Crime')
dfMovies
```

```python
dfMovies=dfMovies[['movieId','Name','year','isAdventure','isAnimation','isChildren','isComedy','isFantasy','isHorror','isThriller','isRomance','isDrama','isCrime']]
dfMovies
```

```python
dfAvgRatings= dfRatings.groupby('movieId', as_index=False).mean()
del dfAvgRatings['userId']
dfAvgRatings['rating']=dfAvgRatings['rating'].round(1)
dfAvgRatings
```

```python
dfNewBoxOffice=dfMovies.merge(dfAvgRatings, on='movieId',how='inner')
dfNewBoxOffice
```

```python
filter1=(dfNewBoxOffice['rating']>3) & (dfNewBoxOffice['rating']<4)
print(filter1)
```

```python
dfMoviesByYear=dfNewBoxOffice.groupby('year').Name.count()
dfMoviesByYear
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(40, 5))
plt.plot(dfNewBoxOffice.groupby('year').Name.count(),c='Green', marker = 'o',ms=5)
plt.legend()

plt.title('Number of movies released per year')
plt.show()
```
